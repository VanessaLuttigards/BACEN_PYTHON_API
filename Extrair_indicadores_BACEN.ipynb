```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import yfinance as yf
import nasdaqdatalink
from functools import reduce
```

##Função para extrair indicadores

```python
def extracao_bcb(codigo, data_inicio, data_fim):
  url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json&dataInicial={}&dataFinal={}'.format(codigo, data_inicio, data_fim)
  df = pd.read_json(url)
  df.set_index('data', inplace= True)
  df.index = pd.to_datetime(df.index, dayfirst=True)
  df.rename(columns={'valor':'Value'},inplace=True)
  return df
```

## Criando Dataframe unificado

```python
# Criando dataframes vazios
confianca_servicos = pd.DataFrame()
confianca_consumidor = pd.DataFrame()
pib = pd.DataFrame()
incc = pd.DataFrame()
igpm = pd.DataFrame()
selic = pd.DataFrame()
ipca = pd.DataFrame()
dolar = pd.DataFrame()
```

```python
# Inserindo os valores nas colunas
confianca_servicos['Confiança_serv'] = extracao_bcb(20339,'01-01-2008','01-03-2024')
confianca_consumidor['Confiança_con']= extracao_bcb(4393,'01-01-2008','01-03-2024')
pib['PIB'] = extracao_bcb(4380,'01-01-2008','01-03-2024')
incc['INCC'] = extracao_bcb(192,'01-01-2008','01-03-2024')
igpm['IGPM']= extracao_bcb(189, '01-01-2008','01-03-2024')
selic['SELIC']= extracao_bcb(432, '01-01-2008','01-03-2024')
ipca['IPCA']= extracao_bcb(13522, '01-01-2008','01-03-2024')
dolar['DOLAR']= extracao_bcb(10813, '01-01-2008','01-03-2024')
```

```python
# Juntando todos os dataframes
indicadores_1 = [confianca_consumidor, confianca_servicos, incc, igpm,pib,selic,ipca,dolar]
indicadores = reduce(lambda x,y : pd.merge(x,y, how = 'inner',on = 'data'),indicadores_1)
```
